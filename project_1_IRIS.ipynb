## Step 2: Seaborn ke Available Datasets Dekhna

---
## Step 3: Dataset Load Karna aur Explore Karna

**Task Requirement:** *Load the dataset using the pandas library*

**Dataset ke baare mein:**
- **150 rows** — phoolon ke samples
- **5 columns** — sepal_length, sepal_width, petal_length, petal_width, species
- **3 species** — Setosa, Versicolor, Virginica
- **Har species mein exactly 50 samples** hain

````python
# sns.load_dataset('iris') → seaborn se iris dataset load karo
# Yeh bilkul pd.read_csv() jaisa hai — bas CSV ki jagah seaborn se aa raha hai
iris_df = sns.load_dataset('iris')

# .head() → Pehli 5 rows dikhao — dataset kaisa dikhta hai
print(' Dataset load ho gaya!')
print('\n Pehli 5 rows (.head()):')
print(iris_df.head())
````

---
## Step 4: Dataset Structure Dekhna — `.shape` `.columns` `.info()`

**Task Requirement:** *Display dataset structure using .shape, .columns, and .head()*

| Function | Kya Batata Hai |
|----------|----------------|
| `.shape` | Kitni rows aur columns hain |
| `.columns` | Har column ka naam |
| `.info()` | Data types aur missing values |
| `.head()` | Pehli 5 rows |

````python
# .shape → (rows, columns) — dataset ka size
print('📐 Dataset Shape (.shape):')
print(iris_df.shape)
print('→ Matlab: 150 phool ke samples, 5 measurements/features')
````

````python
# .columns → Saare column names list mein
print('📋 Column Names (.columns):')
print(iris_df.columns.tolist())
print('→ 4 measurements + 1 species (target) column hai')
````

````python
# .info() → Data types aur missing values check karo
# Non-Null Count = kitni values hain (agar 150 hain → koi missing nahi!)
print('ℹ️ Dataset Info (.info()):')
print(iris_df.info())
print('\n✅ Koi bhi missing value nahi — dataset bilkul clean hai!')
````

---
## Step 5: Basic Statistics Dekhna

**Skill:** *Basic data summarization*

| Function | Kya Batata Hai |
|----------|----------------|
| `value_counts()` | Har species mein kitne samples hain |
| `describe()` | Mean, Min, Max, Average sab ek jagah |

````python
# value_counts() → Har species mein kitne phool hain?
print('🌸 Har Species ka Count:')
print(iris_df['species'].value_counts())
print('\n→ Dataset perfectly balanced hai!')
print('→ Har species mein exactly 50 samples hain!')
````

````python
# .describe() → Statistical summary
# count  = total values
# mean   = average
# std    = kitna spread hai data
# min    = sabse chhoti value
# max    = sabse bari value
# 25/50/75% = quartiles
print('📊 Statistical Summary (.describe()):')
print(iris_df.describe())
````

---
## Step 6: Color Palette Define Karna

Visualizations clear aur attractive banane ke liye
har species ka **alag color** rakha gaya hai:

- 🟣 **Setosa** → Slate Blue
- 🩷 **Versicolor** → Orchid Pink
- 🟤 **Virginica** → Deep Indigo

````python
# Har species ke liye alag color define karo
# Purple/violet theme use kar rahe hain consistency ke liye
colors = {
    'setosa'    : '#7B68EE',   # Medium Slate Blue
    'versicolor': '#DA70D6',   # Orchid Pink
    'virginica' : '#4B0082',   # Deep Indigo
    'background': '#F8F4FF',
    'subtitle'  : '#3C287C',
    'title'     : '#2D1B69'
}

# Seaborn ke liye color list banao
palette = [colors['setosa'], colors['versicolor'], colors['virginica']]

print('✅ Color palette ready!')
print('Setosa=Blue, Versicolor=Pink, Virginica=Indigo')
````

---
## Step 7: Scatter Plot — Sepal Length vs Petal Length

**Task Requirement:** *Scatter plot to analyze relationships between variables*

**Scatter Plot kya hota hai?**
- Har dot = ek phool
- X axis = Sepal Length
- Y axis = Petal Length
- Color = Species (kaun sa phool hai)

**Sawal:** Kya Sepal Length aur Petal Length mein koi relationship hai?
Kya species alag alag dikh ti hai?

````python
# Scatter Plot 1: Sepal Length vs Petal Length
# x='sepal_length' → X axis par sepal length
# y='petal_length' → Y axis par petal length
# hue='species'    → Color se species alag karo
# palette=palette  → Hamare custom colors use karo

plt.figure(figsize=(9, 6))
sns.scatterplot(
    x='sepal_length',
    y='petal_length',
    hue='species',
    data=iris_df,
    palette=palette,
    s=80
)
plt.title('Sepal Length VS Petal Length', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()

print('📌 Insight: Setosa saaf alag dikh raha hai — chota sepal, chota petal!')
print('   Versicolor aur Virginica mein thoda overlap hai.')
````

---
## Step 8: Scatter Plot — Sepal Width vs Petal Width

**Sawal:** Width measurements mein species ka separation kitna clear hai?

````python
# Scatter Plot 2: Sepal Width vs Petal Width
# x='sepal_width' → X axis par sepal width
# y='petal_width' → Y axis par petal width

plt.figure(figsize=(9, 6))
sns.scatterplot(
    x='sepal_width',
    y='petal_width',
    hue='species',
    data=iris_df,
    palette=palette,
    s=80
)
plt.title('Sepal Width VS Petal Width', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()

print('📌 Insight: Petal Width sabse zyada species ko alag karta hai!')
print('   Setosa ka petal width bahut kam (0.1-0.6 cm) hai.')
````

---
## Step 9: Box Plot — Sepal Length by Species

**Task Requirement:** *Box plot to detect outliers and spread of values*

**Box Plot kya dikhata hai?**
```
        |     ← Maximum
    ┌───┤
    │   │  ← 75% values yahan tak
    │───│  ← Median (beech ki value)
    │   │  ← 25% values yahan tak
    └───┤
        |     ← Minimum
        ●     ← Outlier (unusual value)
```

**Sawal:** Kaunsi species mein sepal length zyada hai?

````python
# Box Plot 1: Har species ke liye Sepal Length
# x='species'      → X axis par species naam
# y='sepal_length' → Y axis par sepal length values
# hue='species'    → Har species ka alag color

plt.figure(figsize=(9, 6))
sns.boxplot(
    x='species',
    y='sepal_length',
    data=iris_df,
    hue='species',
    palette=palette,
    width=0.5
)
plt.title('Sepal Length by Species', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.tight_layout()
plt.show()

print('📌 Insight: Virginica ka sepal length sabse zyada hai!')
print('   Setosa mein kuch outliers (dots) hain — unusual measurements.')
````

---
## Step 10: Box Plot — Petal Length by Species

**Sawal:** Petal length mein species ka farq kitna clear hai?

````python
# Box Plot 2: Har species ke liye Petal Length
# Petal length mein species ka difference sabse zyada clear hota hai!

plt.figure(figsize=(9, 6))
sns.boxplot(
    x='species',
    y='petal_length',
    data=iris_df,
    palette=palette,
    width=0.5
)
plt.title('Petal Length by Species', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.tight_layout()
plt.show()

print('📌 Insight: Teen species ka petal length bilkul alag alag hai!')
print('   Setosa: 1-2cm | Versicolor: 3-5cm | Virginica: 4.5-7cm')
print('   Petal Length species pehchaan ne ka sabse acha feature hai!')
````

---
## Step 11: Histogram — Sepal Length Distribution

**Task Requirement:** *Histogram to examine data distribution*

**Histogram kya hota hai?**
- Bars dikhata hai ke ek measurement **kitni baar aayi**
- Zyada lambi bar = us range mein zyada phool
- Chhoti bar = us range mein kam phool

**Sawal:** Dataset mein sepal length sabse zyada kaunsi range mein hai?

````python
# Histogram: Sepal Length ki distribution
# fig, ax = plt.subplots() → Figure aur axis alag alag banao
# bins=20  → Data ko 20 groups mein divide karo
# color    → Bar ka color
# edgecolor → Bars ke beech safed line

fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(
    iris_df['sepal_length'],
    bins=20,
    color='#7B68EE',
    edgecolor='white',
    linewidth=0.8
)
ax.set_title('Sepal Length Distribution', fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Number of Flowers')
plt.tight_layout()
plt.show()

print('📌 Insight: Zyada tar phool 5.5 se 6.5 cm range mein hain!')
print(f'   Average Sepal Length: {iris_df["sepal_length"].mean():.2f} cm')
````
